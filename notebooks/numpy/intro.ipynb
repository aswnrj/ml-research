```python
x = np.array([[3, 2, 5], [1, 1, 2], [2, 3, 1]])
y = np.array([[1, 2, 5], [3, 5, 3], [2, 1, 4]])
```

```python
print(x)
print(y)
print(x+y)
print(np.add(x, y))
print(x-y)
print(np.subtract(x, y))
```

```python
print(x)
print(np.sum(x))
print(np.sum(x, axis=0))
print(np.sum(x, axis=1))
print(x.sum(axis=0))
print(x.min(axis=0))
print(x.max(axis=1))
```
